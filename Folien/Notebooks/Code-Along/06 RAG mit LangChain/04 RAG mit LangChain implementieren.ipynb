
## RAG-Pipeline Übersicht

**Einmalig (Setup)**:
1. Dokumente laden
2. In Chunks aufteilen
3. In ChromaDB speichern (mit Embeddings)

**Bei jeder Anfrage**:
4. Anfrage embedden
5. Ähnlichste Chunks finden (Retrieval)
6. Als Kontext an LLM geben (Augmentation)
7. Antwort generieren (Generation)
