## Was ist Künstliche Intelligenz?

- Wissenschaft und Technik intelligenter Maschinen
- Systeme, die Aufgaben ausführen, die normalerweise menschliche Intelligenz erfordern
- Interdisziplinäres Feld: Informatik, Mathematik, Psychologie, Neurowissenschaften, Philosophie
- Ziel: Maschinen erschaffen, die wahrnehmen, verstehen, lernen und handeln können


## Vier Ansätze zur KI

|              | **Menschlich**     | **Rational**      |
|--------------|--------------------|-------------------|
| **Denken**   | Menschlich denken   | Rational denken  |
| **Handeln**  | Menschlich handeln  | Rational handeln |

- Verschiedene Definitionen und Ziele
- Führen zu unterschiedlichen Forschungsrichtungen

## Menschlich Denken: Kognitive Modellierung

- Ziel: Programme, die wie Menschen denken
- Erfordert Verständnis des menschlichen Geistes
- Methoden:
  - Introspektion
  - Psychologische Experimente
  - Bildgebung des Gehirns
- Kognitionswissenschaft: Verbindung von KI-Modellen und psychologischen Theorien

## Menschlich Handeln: Der Turing-Test

- Vorgeschlagen von Alan Turing (1950)
- Test für intelligentes Verhalten
- Computer muss menschliche Antworten imitieren
- Erforderliche Fähigkeiten:
  - Natürliche Sprachverarbeitung
  - Wissensrepräsentation
  - Automatisches Schlussfolgern
  - Maschinelles Lernen

## Rational Denken: Logik

- Aristoteles: Gesetze des Denkens
- Formale Logik: präzise Notation für Aussagen
- Logistische Tradition in der KI
- Herausforderungen:
  - Informelles Wissen formalisieren
  - Umgang mit Unsicherheit
  - Rechnerische Komplexität


<img src="img/robot-secret-agent.png"

## Rational Handeln: Rationale Agenten

- Agent: Etwas, das handelt
- Rationaler Agent: Handelt, um das beste Ergebnis zu erzielen
- Vorteile gegenüber anderen Ansätzen:
  - Allgemeiner als Logik
  - Wissenschaftlich fundiert (Mathematik, Ökonomie)
  - Erlaubt Annäherungen bei Komplexität

## Geschichte der KI: Anfänge (1943-1956)

- 1943: McCulloch & Pitts - Neuronales Modell
- 1950: Turing - "Computing Machinery and Intelligence"
- 1956: Dartmouth Workshop - Geburt der KI als Feld
  - John McCarthy prägt den Begriff "Künstliche Intelligenz"
  - Teilnehmer: Minsky, Simon, Newell, Shannon

## KI-Gebiete

- **Maschinelles Lernen**: Systeme, die aus Erfahrung lernen
- **Computer Vision**: Bildverstehen und -verarbeitung
- **Natürliche Sprachverarbeitung**: Textverstehen und -generierung
- **Robotik**: Physische Agenten in der realen Welt
- **Wissensrepräsentation**: Modellierung von Wissen


## KI-Gebiete

- **Planung**: Sequenzen von Aktionen zur Zielerreichung
- **Expertensysteme**: Spezialisiertes Domänenwissen
- **Genetische Algorithmen**: Evolutionäre Optimierung
- **Schlussfolgern unter Unsicherheit**: Umgang mit Vagheit, Fuzzy-Logik
- **Multi-Agenten-Systeme**: Koordination mehrerer Agenten


## Stand der Technik: Erfolge

- **Spiele**: Übermenschliche Leistung in Schach, Go, Poker
- **Spracherkennung**: Nahezu menschliche Genauigkeit
- **Übersetzung**: Hochwertige automatische Übersetzung
- **Bilderkennung**: Übertrifft Menschen bei bestimmten Aufgaben
- **Medizin**: Diagnose von Krankheiten, Medikamentenentdeckung
- **Autonomes Fahren**: Fortgeschrittene Assistenzsysteme


## Stand der Technik: Generative KI

- **Large Language Models (LLMs)**
  - GPT-5, Claude, Gemini
  - Textgenerierung, Programmierung, Analyse
- **Bildgenerierung**
  - DALL-E 3, Midjourney, Stable Diffusion
  - Fotorealistische und künstlerische Bilder
- **Multimodale Modelle**
  - Verarbeitung von Text, Bild, Audio, Video
  - Vision-Language Models


## Stand der Technik: Herausforderungen

- **Erklärbarkeit**: Black-Box-Modelle schwer zu interpretieren
- **Robustheit**: Anfällig für "bösartige" Beispiele
- **Generalisierung**: Transfer auf neue Domänen schwierig
- **Ethik & Bias**: Fairness und Diskriminierung
- **Energieverbrauch**: Training großer Modelle sehr ressourcenintensiv
- **Halluzinationen**: KI generiert plausible aber falsche Informationen


## Offene Probleme

- **Allgemeine KI (AGI)**: Noch weit entfernt
- **Bewusstsein**: Können Maschinen bewusst sein?
- **Gesunder Menschenverstand**: Schwierig zu formalisieren
- **Kausalität**: Verstehen von Ursache und Wirkung
- **Langzeit-Autonomie**: Selbstständiges Lernen und Anpassen
- **Mensch-KI-Kollaboration**: Optimale Zusammenarbeit


## Zusammenfassung

- KI: Wissenschaft intelligenter Agenten
- Vier Ansätze: menschlich/rational × denken/handeln
- Moderne KI fokussiert auf rationale Agenten
- Beeindruckende Fortschritte in vielen Bereichen
- Generative KI revolutioniert viele Anwendungen
- Wichtige Herausforderungen bleiben bestehen
